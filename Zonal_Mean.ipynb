```python
ds = xr.open_dataset("CMIP.IPSL.IPSL-CM6A-LR.historical.Omon.gn.nc")
```

```python
zonal = ds.sel(time=slice("1980", "2010")).mean(dim="time")
```

```python
zonal.so.plot.contourf(x="lat", levels=30, cmap="RdYlBu_r", vmin=33.4, vmax=36.4)
plt.ylim(0, 2000)
plt.gca().invert_yaxis()
plt.show()
```
